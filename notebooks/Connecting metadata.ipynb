```python
def copy_recordings_to_parent_folder(bucket_name, source_prefix, destination_prefix):
    # Initialize S3 client
    s3 = boto3.client('s3')

    # List objects in the source prefix (recursive)
    response = s3.list_objects_v2(Bucket=bucket_name, Prefix=source_prefix)

    # Iterate through objects in the source prefix
    for obj in response.get('Contents', []):
        # Get the object key
        object_key = obj['Key']

        # Extract the folder structure from the object key
        folders = object_key.split('/')
        
        # Skip if the object is not in the expected folder structure
        if len(folders) != 5:
            continue

        # Extract the year, month, and day from the object key
        year, month, day = folders[1], folders[2], folders[3]

        # Construct the new object key in the parent folder
        new_object_key = f"{destination_prefix}/{year}/{month}/{day}/{folders[-1]}"

        # Copy the object to the new location
        s3.copy_object(CopySource={'Bucket': bucket_name, 'Key': object_key},
                       Bucket=bucket_name, Key=new_object_key)

        print(f'Copied {object_key} to {new_object_key}')
```

```python
bucket_name = '2307-01-acoustic-loggers-for-leak-detection-a'
source_prefix = 'Unstructured audio files/'
destination_prefix = 'Unsructured audio files'

copy_recordings_to_parent_folder(bucket_name, source_prefix, destination_prefix)
```

```python
import boto3

def copy_audio_files_to_destination(bucket_name, source_prefix, destination_prefix):
    # Initialize S3 client
    s3 = boto3.client('s3')

    # List objects in the source prefix (recursive)
    response = s3.list_objects_v2(Bucket=bucket_name, Prefix=source_prefix)

    # Iterate through objects in the source prefix
    for obj in response.get('Contents', []):
        # Get the object key
        object_key = obj['Key']

        # Check if the object is a folder
        if object_key.endswith('/'):
            # Recursively call the function for subfolders
            copy_audio_files_to_destination(bucket_name, object_key, destination_prefix)
        else:
            # Check if the object is an audio file (adjust condition as needed)
            if object_key.endswith('.wav') or object_key.endswith('.mp3'):
                # Construct the new object key in the destination folder
                destination_object_key = f"{destination_prefix}/{object_key.split('/')[-1]}"

                # Copy the object to the destination folder
                s3.copy_object(CopySource={'Bucket': bucket_name, 'Key': object_key},
                               Bucket=bucket_name, Key=destination_object_key)

                print(f'Copied {object_key} to {destination_object_key}')

# Example usage
bucket_name = '2307-01-acoustic-loggers-for-leak-detection-a'
source_prefix = 'Unstructured audio files/'
destination_prefix = 'All_Audios'

copy_audio_files_to_destination(bucket_name, source_prefix, destination_prefix)
```
